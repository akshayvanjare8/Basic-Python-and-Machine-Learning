#### Convert Data Normalisation and Perform Multiple Linear Regression

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x = bodyfat[["Density","Weight","Chest","Abdomen","Hip","Thigh","Knee","Biceps"]]
x.head()
```

```python
y = bodyfat[["bodyfat"]]
y.head()
```

```python
m = MinMaxScaler()

m.fit(x)

x_tran = m.transform(x)

sns.boxplot(data=x_tran)
plt.show()
```

```python
m = MinMaxScaler()

m.fit(y)

y_tran = m.transform(y)

sns.boxplot(data=y_tran)
plt.show()
```

```python
model = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_tran, y_tran, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pd.DataFrame(model.coef_.flatten(), x.columns, columns = ['Coeff'])
```

```python
ypred = model.predict(x_test)
```

```python
error = y_test - ypred
```

```python
error.mean()
```

```python
sns.histplot(error, kde = True)
plt.show()
```

```python
plt.scatter(ypred, error)
plt.show()
```
