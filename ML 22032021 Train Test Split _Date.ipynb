```python
# Train Test Split DataSet
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.head()
```

```python
x = data.drop(columns ="bodyfat")
```

```python
y = data["bodyfat"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
#print(x_train)
```

```python
#print(x_test)
```

```python
#print(y_train)
```

```python
#print(y_test)
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test) # checking accuracy
```

```python
#Conclusion
#We have seen how we can split up the dataset into training and testing parts to pass it to the fit function. 
#Also, we have used the linear-regression algorithm to predict the bodyfat of the body on the basis of other independent variables.
```

```python
model = LinearRegression().fit(x_train, y_train)
print("Intercept :", model.intercept_)

print ("Coefficients :", model.coef_)
#
```

```python
model.score(x_train, y_train) #Accuracy and Value of R_Square
```

```python
model.score(x_test, y_test) #Accuracy and Valur of R_Square
```

```python
# Conclusion :
#The higher the R value, the better the fit. In this case, the testing data yields a slightly higher coefficient. 
#However, the R calculated with test data is an unbiased measure of your models prediction performance.
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)

print("R_suare or Accuracy for Train :", model.score(x_train, y_train))

print("R_square or Accuracy for Test :",model.score(x_test, y_test))
```

```python
# Boosting Regressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(random_state=0).fit(x_train, y_train)

print("R_suare or Accuracy for Train :", model.score(x_train, y_train))

print("R_square or Accuracy for Test :",model.score(x_test, y_test))
```

```python
# Random Forest Regressor
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=0).fit(x_train, y_train)

print("R_suare or Accuracy for Train :", model.score(x_train, y_train))

print("R_square or Accuracy for Test :",model.score(x_test, y_test))
```
