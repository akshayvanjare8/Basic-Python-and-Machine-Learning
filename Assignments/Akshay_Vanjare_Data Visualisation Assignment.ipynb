# 2. COVID Dataset

```python
covid=pd.read_csv("covid_19_india.csv")

covid.head()
```

```python
covid.info()
```

```python
covid.isnull().sum()
```

```python
sns.pairplot(data=covid)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid['Confirmed'])
plt.title('Confirmed Cases')
plt.show()
```

Conclusion : From the above graph, we can say that the shape of confirmed cases are monotonically increasing. 

```python
plt.figure(figsize=(10,5))
plt.plot(covid['Deaths'])
plt.title('Deaths Cases')
plt.show()
```

Conclusion : From the above graph, we can say that the shape of death cases are monotonically increasing. 

```python
x=covid['Confirmed'].sum()
y=covid['Cured'].sum()
z=covid['Deaths'].sum()
active=x-y

print('Total Confirmed cases =', x)
print('Total Cured cases =', y)
print('Total Death cases =', z)
print('Total Active cases =', x)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=['Confirmed','Cured','Deaths','Active'], y=[x,y,z,active])
plt.show()
```

Conclusion : From the above graph, we can say that the number of confirmed cases are greater than as compare to number of cured, Deaths and active cases.
    Also we can see that number of deaths cases are less than others.

```python
# top 10 containment states

covid_df1 = covid.tail(20)
```

```python
covid_df2 = covid_df1.sort_values(by='Confirmed', ascending=False).head(10)

covid_df2
```

```python
plt.figure(figsize=(20,10))
plt.bar(covid_df2['State/UnionTerritory'],height=covid_df2['Confirmed'],color='blue')

plt.show()
```

Conclusion : From the above graph, we can say that confirmed cases are maximum in maharashtra state and then next in karnataka state. In other side Madhya Pradesh state have very low confirmed cases.

```python
covid_df2['Fatality-Ratio'] = covid_df2['Deaths']/covid_df2['Confirmed']

plt.figure(figsize=(20,10))
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=covid_df2,x='State/UnionTerritory',y='Fatality-Ratio',color='orange')
plt.show()
```

Conclusion : From the above graph, we can say that fatality ratio are maximum in maharashtra state and then next in west bengal state. In other side kerala state have very low fatality ratio.

```python
covid_df2['Cure-Ratio'] = covid_df2['Cured']/covid_df2['Confirmed']

plt.figure(figsize=(20,10))
plt.title("Cure Ratio")
sns.pointplot(data=covid_df2,x='State/UnionTerritory',y='Cure-Ratio',color='blue')
plt.show()
```

Conclusion : From the above graph, we can say that cure ratio are maximum in odisa state and then next in tamilnadu state. In other side kerala state have very low cure ratio.

```python
#most affected state i.e Maharastra

mh_df=covid[covid['State/UnionTerritory']=='Maharashtra']
mh_df
```

```python
mh_month=mh_df[mh_df['Date'].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
mh_month
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=mh_month,x='Date',y='Confirmed',color='Blue')
sns.lineplot(data=mh_month,x='Date',y='Cured',color='Green')
sns.lineplot(data=mh_month,x='Date',y='Deaths',color='Red')
plt.show()
```

Conclusion : From the above graph, from 09/07/20 confirm cases and deaths cases increases.

```python
# line plot time series of all states data

state_month=covid[covid['Date'].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
state_month
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=state_month,x='Date',y='Confirmed',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Confirmed Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=state_month,x='Date',y='Deaths',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Death Cases")
plt.show()
```

Conclusion : from above two graphs, the Maharashtra state has most death cases and confirmed cases
