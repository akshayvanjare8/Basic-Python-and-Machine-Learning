#### Missing Values Check

```python
data.isna().sum()
```

```python
data["Street Number Suffix"] = data["Street Number Suffix"].fillna("U")

data["Street Suffix"] = data["Street Suffix"].fillna("U")

data["Unit"] = data["Unit"].fillna(data["Unit"].mean())

data["Unit Suffix"] = data["Unit Suffix"].fillna("U")

data["Description"] = data["Description"].fillna("U")

data["Issued Date"] = data["Issued Date"].fillna("U")

data["Completed Date"] = data["Completed Date"].fillna("U")

data["First Construction Document Date"] = data["First Construction Document Date"].fillna("U")

data["Structural Notification"] = data["Structural Notification"].fillna("U")

data["Number of Existing Stories"] = data["Number of Existing Stories"].fillna(data["Number of Existing Stories"].mean())

data["Number of Proposed Stories"] = data["Number of Proposed Stories"].fillna(data["Number of Proposed Stories"].mean())

data["Voluntary Soft-Story Retrofit"] = data["Voluntary Soft-Story Retrofit"].fillna("U")

data["Fire Only Permit"] = data["Fire Only Permit"].fillna("U")

data["Permit Expiration Date"] = data["Permit Expiration Date"].fillna("U")

data["Estimated Cost"] = data["Estimated Cost"].fillna(data["Estimated Cost"].mean())

data["Revised Cost"] = data["Revised Cost"].fillna(data["Revised Cost"].mean())

data["Existing Use"] = data["Existing Use"].fillna("U")

data["Existing Units"] = data["Existing Units"].fillna(data["Existing Units"].mean())

data["Proposed Use"] = data["Proposed Use"].fillna("U")

data["Proposed Units"] = data["Proposed Units"].fillna(data["Proposed Units"].mean())

data["Plansets"] = data["Plansets"].fillna(data["Plansets"].mean())

data["TIDF Compliance"] = data["TIDF Compliance"].fillna("U")

#data[""] = data[""].fillna(data[""].mean())

data["Existing Construction Type"] = data["Existing Construction Type"].fillna(data["Existing Construction Type"].mean())

data["Existing Construction Type Description"] = data["Existing Construction Type Description"].fillna("U")

data["Proposed Construction Type"] = data["Proposed Construction Type"].fillna(data["Proposed Construction Type"].mean())

data["Proposed Construction Type Description"] = data["Proposed Construction Type Description"].fillna("U")

data["Site Permit"] = data["Site Permit"].fillna("U")

data["Supervisor District"] = data["Supervisor District"].fillna(data["Supervisor District"].mean())

data["Neighborhoods - Analysis Boundaries"] = data["Neighborhoods - Analysis Boundaries"].fillna("U")

data["Zipcode"] = data["Zipcode"].fillna("U")

data["Location"] = data["Location"].fillna("U")

#data[""] = data[""].fillna(data[""].mean())

```

#### Scaling And Normalisation

```python
data.head()
```

```python
df_num = data[['Unit','Number of Existing Stories', 'Number of Proposed Stories',
       'Estimated Cost', 'Revised Cost', 'Existing Units', 'Proposed Units',
       'Plansets', 'Existing Construction Type', 'Proposed Construction Type',
       'Supervisor District']]
df_num.head()
```

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on  data
norm = MinMaxScaler()

norm.fit(df_num)

# transform data
data_norm = norm.transform(df_num)
```

```python
data_norm = pd.DataFrame(data_norm)
data_norm.head()
```

```python
# data normalization with sklearn
from sklearn.preprocessing import StandardScaler

# fit scaler on  data
ss = StandardScaler()
ss.fit(df_num)

# transform data
data_ss = ss.transform(df_num)
```

```python
data_ss = pd.DataFrame(data_ss)
data_ss.head()
```
