```python
df.isna().sum() 
```

```python
# DataFrame to List 

trans = [] 
for i in range(df.shape[0]):
    sub = []
    for j in range(df.shape[1]):
        if str(df.values[i,j]) != '0':
            sub.append(df.values[i,j]) 
        trans.append(sub)
```

```python
trans[0] 
```

```python
# List to List 

trans = [] 

for i in range(df.shape[0]):
    trans.append([str(df.values[i,j]) for j in range(df.shape[1]) if str(df.values[i,j]) != '0'])
```

```python
trans[0] 
```

```python
from apyori import apriori
```

```python
rules = apriori(trans, min_support = 0.003, min_confidance = 0.2, min_lift = 3, min_length = 2)
```

```python
rules 
```

```python
df_rules = pd.DataFrame(list(rules))
df_rules.head() 
```

```python
df_rules["ordered_statistics"] 
```

```python
lhs = []
rhs = []
conf = [] 
lift = []

for i in range(df_rules.shape[0]):
    order_stat = df_rules["ordered_statistics"][i][0]
    lhs.append(list(order_stat[0]))
    rhs.append(order_stat[1]) 
    conf.append(order_stat[2])
    lift.append(order_stat[3]) 
```

```python
final_report = pd.DataFrame({"LHS":lhs, "RHS":rhs, "Support":df_rules.support, "Confidance":conf, "Lift":lift})
final_report 
```

```python
final_report.sort_values(by = "Confidance")
```
