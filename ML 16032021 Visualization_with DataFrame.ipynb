### Titanic Data set Visualisation

```python
url="https://raw.githubusercontent.com/akshayvanjare8/datset/main/train.csv"

titanic=pd.read_csv(url)
titanic.head()
```

```python
titanic.info()
```

```python
#check missing values

titanic.isna().sum()
```

```python
# fill missing values 

titanic['Age']=titanic['Age'].fillna(titanic['Age'].mean)

titanic['Embarked']=titanic['Embarked'].fillna(titanic['Embarked'].mode)

titanic['Cabin']=titanic['Cabin'].fillna("U")
```

```python
titanic.isna().sum()
```

```python
titanic['Age'] = pd.to_numeric(titanic['Age'], errors = 'coerce')
titanic['Age'] = titanic.Age.astype(int)
```

```python
sns.catplot(x="Sex", data= titanic, kind="count")
plt.show()
```

Conclusion : In this Dataset count of male is greater than count of female

```python
sns.catplot(x="Pclass", data= titanic, kind="count")
plt.show()
```

Conclusion : passengers of 3rd class is greater than othe class

```python
sns.catplot(x="Pclass", data= titanic, kind="count", hue="Sex")
plt.show()
```

Conclusion : In all class, male passenger is greater than female passengers.
    and in 3rd class male passangers are double than female passangers

```python
# crete new coulmn for children

def check_age(data):
    Age, Gender=data 
    if Age<16:
        return "child"
    else:
        return Gender 
    
titanic["People"]=titanic[["Age","Sex"]].apply(check_age, axis=1)
print(titanic["People"])
```

```python
sns.factorplot(x="People", data=titanic, kind="count")
plt.show()
```

```python
sns.factorplot(x="Pclass", data=titanic,kind="count", hue="People")
plt.show()
```

```python
def check_family(data):
    Parch, SibSp= data 
    c=Parch+SibSp 
    if c>0:
        return "without family"
    else:
        return "with family"
titanic["Family"]=titanic[["Parch","SibSp"]].apply(check_family,axis=1)
#
```

```python
sns.catplot(x="Family", data=titanic, kind="count")
plt.show()
```

```python
sns.catplot(x="Family", data=titanic, kind="count", hue="People")
plt.show()
```
