# Data Cleaning

### Check Missing Values 

```python
print(train.isna())
```

```python
print(train.isna().sum())
```

#### Using Dropna

```python
train=train.dropna() #remove missing values using nes

print(train.isnull().sum()) 

print(train.tail())
```

#### Using Mean Median Mode and U

```python
# fill missing values by mean 

mean=train['Age'].mean()
#print(mean)

train['Age']=train['Age'].fillna(mean)

# fill missing character by U 

train['Name']=train['Name'].fillna('U')

train['Cabin']=train['Cabin'].fillna('U')

# fill missing values by mode 

mode=train['Embarked'].mode()[0]
#print(mode)

train['Embarked']=train['Embarked'].fillna(mode)

# fill missing values by median 
#median=train['column name'].median()[0]
#print(median)

#train['column name']=train['column name'].fillna(mode)
```

#### using forward fill and backward fill

```python
#forward fill 

#methods=>pad,ffill

df=df.fillna(method="pad") #syntax
print(df)

#backward fill 

#methods=>bfill 

df=df.fillna(method="bfill") #syntax
print(df)
```
