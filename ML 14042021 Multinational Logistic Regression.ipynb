```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
data.info()
```

```python
data["species"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

data["label_species"] = le.fit_transform(data["species"])
data.head()
```

```python
data["label_species"].value_counts()
```

```python
x = data.drop(columns = ["species", "label_species"])
y = data[["label_species"]]
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression(multi_class = "multinomial")
```

```python
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
from sklearn import metrics 

acc = metrics.accuracy_score(y, y_pred)

print(acc)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, y_pred)
print(cm)
```

```python
sns.heatmap(cm, annot = True)
plt.show()
```

```python
print(metrics.classification_report(y, y_pred))
```

### Hyper Parameter Tuning

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
x = df.drop(columns = ["Outcome"])

y = df[["Outcome"]]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
help(LogisticRegression)
```

```python
#solver, penalty, c
```

```python
model = LogisticRegression()
solver = ["newton-cg","lbfgs","liblinear","sag","saga"]
c = [100,10,1,0.1,0.01,0.001]
penalty = ["l1","l2","elsticnet"]
grid = {"solver":solver, "penalty":penalty, "C":c}
grid
```

```python
# cv cross validator
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
kfolds_cv = RepeatedStratifiedKFold(n_splits = 10, n_repeats = 5, random_state = 1)
```

```python
# Grid search cv
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_cv = GridSearchCV(estimator = model, param_grid = grid, cv = kfolds_cv, scoring = "accuracy")
```

```python
res = grid_cv.fit(x,y)
```

```python
res.best_params_
```

```python
res.best_score_
```

```python
res.best_estimator_
```
