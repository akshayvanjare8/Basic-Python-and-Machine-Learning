# Visualisation

```python
import matplotlib.pyplot as plt 
#import seaborn
```

### line plot

```python
x = [1,2,3,4,5]
y = [7,9,11,13,15]
```

```python
plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y, marker="*", color='r', label="myline")
plt.legend()
plt.show()
```

```python
plt.plot(x,y, marker="*", color='r', label="myline1")
plt.legend()
plt.show()
plt.plot(x,y, marker="*", color='r', label="myline2")
plt.legend()
plt.show()
```

```python
#title

plt.plot(x,y, marker="*", color='r', label="myline")
plt.legend()
plt.title("my first line plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```

```python
#ticks 

plt.plot(x,y, marker="*", color='r', label="myline")
plt.legend()
plt.title("my first line plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([1,3,5])
plt.yticks([1,3,5])
plt.show()
```

```python
plt.plot(x,y, marker="*", color='r', label="myline", linestyle="--", markerfacecolor="b")
plt.legend()
plt.title("my first line plot")
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.xticks([1,3,5],["day1","day2","day3"])
plt.yticks([1,3,5])
plt.show()
```

### scatter plot

```python
plt.scatter(x,y, color='r', s=100)
plt.show()
```

```python
plt.scatter(x,y, color=["b","r","g","k","b"], s=100)
plt.show()
```

### visualisation of DataFrame

```python
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/akshayvanjare8/datset/main/Teams.csv"

team = pd.read_csv(url)

print(team.head())
print(team.tail())
```

```python
teams = team.groupby("Team")

teams_sum = teams ["Points"].sum()
print(teams_sum)
```

```python
plt.bar(range(4),teams_sum)
plt.xticks(range(4), teams_sum.index)
plt.show()
```
